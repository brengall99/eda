```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

First let's get a quick look at the data, and how it is formatted

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.set_theme()
```

```python
sns.countplot(x='type', data=df)
```

```python
df.fillna('Unknown', inplace=True)
```

```python
df.isna().sum()
```

```python
df_expanded = df['country'].str.split(',').explode().str.strip()

country_counts = df_expanded.value_counts()
country_counts = country_counts.head(20)
plt.figure(figsize=(15, 16))
country_counts.plot(kind='pie', autopct='%1.1f%%', startangle=60)
plt.title('Top 20 Distribution of Movies by Country')
```

```python
df['release_year'].value_counts()
```

```python
# Filter the DataFrame to include only rows where release_year >= 1980
filtered_df = df[df['release_year'] >= 1990]
plt.figure(figsize=(10, 6))
sns.countplot(x='release_year', data=filtered_df, palette='Blues')
plt.xticks(rotation=90)
```

```python
df['rating'].value_counts().head(7).plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(8, 8))
plt.title('Distribution of Ratings')
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.extract(r'(\d+)').astype(float)
movies = movies[movies['duration'] < 220]

plt.figure(figsize=(12, 6))
sns.histplot(movies['duration'], color='red', kde=False, label='Movies', bins=30)

plt.title('Duration Analysis of Netflix Movies', fontsize=16)
plt.xlabel('Duration (minutes)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
shows = df[df['type'] == 'TV Show'].copy()
shows['duration'] = shows['duration'].str.extract(r'(\d+)').astype(float)

plt.figure(figsize=(12, 6))
sns.histplot(shows['duration'], color='blue', kde=False, label='Shows', bins=30)
plt.xticks(ticks=range(1, 16))

plt.title('Duration Analysis of Netflix Shows', fontsize=16)
plt.xlabel('Duration (seasons)')
plt.ylabel('Count', fontsize=12)
plt.show()
```
