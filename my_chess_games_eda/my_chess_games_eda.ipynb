```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.describe().round(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['date'] = df['date'].str.strip()  # Remove any extra spaces
df['date'] = pd.to_datetime(df['date'], format='%Y.%m.%d')
```

```python
sns.lineplot(x='date', y='userRating', data=df)

# Show the plot
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='date', y='userRating', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df_filtered = df[df['date'].dt.year == 2018]
```
